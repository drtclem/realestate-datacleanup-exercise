#### Exercise 08. Plot the histogram of prices for the population (level5 column) of "Arroyomolinos (Madrid)" and explain what you observe (★★☆)

Print the histogram of the prices and write in the Markdown cell a brief analysis about the plot.

```python
ds_arroyo_madrid['price'].plot(kind='hist' , bins = 20, edgecolor='black', title = 'Histogram of Price')

plt.xlabel('Price')
plt.ylabel('Frequency')
```

The histogram above reveals a fairly normal distribution of prices. That is, the most common prices are in closer to the middle of the entire range. However, there are some prices that are less common in the 'middle', this may suggest that pricing trends have specific associations with potential buyers. For example, 250 and 300 may be standard expectations for buyers, leaving values between 250 and 300 less likely to be used. Potentially a tiered expectation or 'slotting' from the market. 


#### Exercise 09. Are the average prices of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about them.

```python
ds_valdemorillo = ds[ds['level5'] == 'Valdemorillo']
ds_galapagar = ds[ds['level5']== 'Galapagar']


average_price_vald = ds_valdemorillo['price'].mean()
average_price_gala = ds_galapagar['price'].mean()

print(round(average_price_vald, 2))
print(round(average_price_gala,2))

```

#### Exercise 10. Are the average prices per square meter (price/m2) of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about it.

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
ds.plot(kind='scatter', x='price', y='surface', title='Graph of Surface Are versus Price')

plt.yticks([0, 2000, 250])
plt.xlabel('Price per Millions')
plt.ylabel('Surface Area in M^2')
x = ds['price']
y=ds['surface']
m, b = np.polyfit(x,y,1)
plt.plot(x,m*x+b, color = 'black', label =f'Best Fit Line')

plt.show()

```
